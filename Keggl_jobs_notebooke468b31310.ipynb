```python
df = pd.read_csv(r"/kaggle/input/heart-disease-prediction-using-logistic-regression/framingham.csv")
print(df.columns)
df.head(5)
```

```python
df.describe()
```

```python
from sklearn.model_selection import StratifiedShuffleSplit
split = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42)
for train_index, test_index in split.split(df, df["prevalentStroke"]):
 strat_train_set = df.loc[train_index]
 strat_test_set = df.loc[test_index]
```

```python
strat_train_set.info()
```

```python
data_label = strat_train_set['prevalentStroke'].copy()
strat_train_set.drop('prevalentStroke', axis = 1, inplace=True)
strat_train_set.head(2)
```

```python
# Importantttttttttttt
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
num_pipeline = Pipeline([
 ('imputer', SimpleImputer(strategy="median")), 
#  ('std_scaler', StandardScaler()),
 ])
strat_train_set_prepped = num_pipeline.fit_transform(strat_train_set)
strat_train_set_df = pd.DataFrame(strat_train_set_prepped, columns = strat_train_set.columns)
strat_train_set_df.info()
```

```python
from sklearn.linear_model import LogisticRegression

logreg = LogisticRegression()
logreg.fit(strat_train_set_prepped,data_label)
```

```python
test_data = strat_test_set.iloc[:5]
result = test_data['prevalentStroke'].copy()
test_data.drop('prevalentStroke',axis = 1, inplace = True)
test_data.head()
```
