Linear Regressor

```python
import pandas as pd
from sklearn.linear_model import LinearRegression

housing_df = pd.read_csv(r'D:\python projects\ML_Self_Prac\Housing.csv')
test_df = housing_df.iloc[:490]
test_df.info()

linear_reg = LinearRegression()
linear_reg.fit(test_df[['area','bedrooms','bathrooms','stories']], test_df.price)
# linear_reg.predict()
```

```python
test_df = housing_df[490:]
test_data = test_df[['area','bedrooms','bathrooms','stories']].to_numpy()
linear_reg.predict(test_data)
```

```python
linear_reg.coef_
```

```python
linear_reg.intercept_
```

```python
import matplotlib.pyplot as plt
plt.scatter(test_df['area'], test_df['price'], marker='*')
plt.plot(test_df.area, linear_reg.predict(test_df[['area','bedrooms','bathrooms','stories']]), color = 'blue')
```
