```python
import pandas as pd

housing_data = pd.read_csv(r"D:\Version_Controlled\Version_Controlled\Practice\ML_Self_Prac\Housing.csv")
housing_data.describe()
```

```python
housing_data.head(5)
```

Nominal data conversion

```python
dummies = pd.get_dummies(housing_data[['furnishingstatus']])
dummies
```

```python
mergged_df = pd.concat([housing_data,dummies], axis="columns")
mergged_df.head(5)
```

Nominal data conversion using sklearn

```python
from sklearn.preprocessing import OrdinalEncoder
onehotencoder = OrdinalEncoder()
categorical_df = onehotencoder.fit_transform(housing_data[['furnishingstatus']])
housing_cat_1hotencoded_df = pd.DataFrame(categorical_df, columns=['furnishingstatus'])
new_mergged_df = pd.concat([housing_data,housing_cat_1hotencoded_df], axis="columns")
new_mergged_df.head(5)
```

Using Labelencoder

```python
from sklearn.preprocessing import LabelEncoder
lablencoder = LabelEncoder()
label_data = lablencoder.fit_transform(housing_data['furnishingstatus'])
housing_label_df = pd.DataFrame(label_data, columns = ['furnishingstatus'])
label_mergged_df = pd.concat([housing_data,housing_label_df], axis="columns")
label_mergged_df.head(5)

















































```
